```python
def get_tickers():
    companies=pd.read_csv('https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/data/module-1/companies.csv')
    list_of_tickers=companies.Ticker.to_list()
    print('Retrieved', len(list_of_tickers), 'tickers')
    return list_of_tickers
```

```python
def get_prices(ticker):
    print('Retrieving data for', ticker)
    prices=quandl.get('WIKI/'+list_of_tickers[0])['Adj. Close'].reset_index()
    prices['Ticker']=ticker
    return prices
```

```python
get_tickers()
```

```python
get_prices(list_of_tickers[11])
```

```python
def acquisition():
    tickers=get_tickers()
    data=[get_prices(ticker) for ticker in tickers]
    df=pd.concat(data, sort=True)
    print('Done, fin!')
    return df
```

```python
df=acquisition()
```

```python
df.sample(10)
```

```python
df.pivot_table(index='Date',columns='Ticker', values='Adj. Close')
```

```python
def concat_pivot(df, rows='Date', columns='Ticker', data='Adj. Close'):
    return df.pivot_table(index=rows, columns=columns, values=data)             
```

```python
import matplotlib.pyplot as plt
```

```python
def compute_pct_change(df):
    return df.pct_change()
```

```python
def plot(df):
    df.plot()
    plt.show()
```

"""We have prepared 4 functions. Now we want to form a pipeline of functions that will work together. Let's see 3 ways of doing this."""

```python
plot(compute_pct_change(concat_pivot(acquisition())))
#This is the worst way of preparing our functions
```

```python
# better, acceptable
df=acquisition()
df=concat_pivot(df)
df=compute_pct_change(df)
plot(df)
```

```python
#the best way
(acquisition()
 .pipe(concat_pivot)
 .pipe(compute_pct_change)
 .pipe(plot))
```
